# Problem - 4: Matrix Operations:

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
A_inverse = np.linalg.inv(A)
A_transpose = np.transpose(A)
B_transpose = np.transpose(B)
I = np.eye(2)
```

```python
if np.allclose((A @ A_inverse), I):
  print("A.A−1 = I. proved")
```

```python
if not np.allclose((A @ B), (B @ A)):
  print("AB ̸= BA. proved")
```

```python
if np.allclose(np.transpose(A @ B), (B_transpose @ A_transpose)):
  print("(Transpose of (AB)) == (transpose of B * Transpose of A) proved")
```

# **4.2 Experiment: How Fast is Numpy?**

```python
import time
import random
```

```python
size = 1_000_000

list1 = list(range(size))
list2 = list(range(size))

arr1 = np.array(list1)
arr2 = np.array(list2)
```

```python
start = time.time()

result_list = [list1[i] + list2[i] for i in range(size)]

end = time.time()
print("Time taken using Python lists for element-wise addition:", end - start, "seconds")
```

```python
start = time.time()

result_array = arr1 + arr2

end = time.time()
print("Time taken using NumPy arrays for element-wise addition:", end - start, "seconds")
```

```python
start = time.time()

result_list = [list1[i] * list2[i] for i in range(size)]

end = time.time()
print("Time taken using Python lists for element-wise multiplication:", end - start, "seconds")
```

```python
start = time.time()

result_array = arr1 * arr2

end = time.time()
print("Time taken using NumPy arrays for element-wise multiplication:", end - start, "seconds")
```

```python
start = time.time()

result_list = sum(list1[i] * list2[i] for i in range(size))

end = time.time()
print("Time taken using Python lists for dot product:", end - start, "seconds")
```

```python
start = time.time()

result_array = np.dot(arr1, arr2)

end = time.time()
print("Time taken using NumPy arrays for dot product:", end - start, "seconds")
```

```python
size = 1000

matrix1 = [[random.randint(0, 10) for _ in range(size)] for _ in range(size)]
matrix2 = [[random.randint(0, 10) for _ in range(size)] for _ in range(size)]

arr1 = np.array(matrix1)
arr2 = np.array(matrix2)
```

```python
start = time.time()

result_list = [[0]*size for _ in range(size)]

for i in range(size):
    for j in range(size):
        for k in range(size):
            result_list[i][j] += matrix1[i][k] * matrix2[k][j]

end = time.time()
print("Time taken using Python lists for matrix multiplication:", end - start, "seconds")
```

```python
start = time.time()

result_array = arr1 @ arr2

end = time.time()
print("Time taken using NumPy arrays for matrix multiplication:", end - start, "seconds")
```
